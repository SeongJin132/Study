```python
## 관심있는 주식의 일별시세 페이지 알아내기
## 일별시세 페이지의 [Headers] 탭에서 웹 요청 주소(Request URL) 알애내기
# https://finance.naver.com/item/sise_day.naver?code=005930&page=2
```

```python
## user-agent 값 알아내기
# User-Agent:Mozilla/5.0
```

```python
## HTTP 요청을 보낼 때 함께 전송하는 데이터 페이로드(Payload) 알아내기(주식코드와 페이지 정보)
# code: 005930
# page: 2
```

```python
# 웹 스크래핑 가능 유무 확인을 위한 모듈 탑재(requests)
import requests
```

```python
# 웹 스크래핑 대상 주소 저장
url = 'https://finance.naver.com/item/sise_day.naver?code=005930&page=2'
```

```python
# HTTP 요청을 대상 주소로 보내고, HTTP 응답 받아서 저장하기(requests.get())
page = requests.get(url)
```

```python
# 응답 내용 확인하기
print(page.text)
```

```python
## 웹브라우저에서 보내는 요청으로 헤더(headers) 설정위해
## 헤더 정보 설정하기(딕셔너리 형태로)
my_headers = {'User-Agent':'Mozilla/5.0'}
```

```python
# 요청 헤더 설정하여 HTTP 응답 받아오기
data = requests.get(url,headers =my_headers)
```

```python
# 응답 내용 확인하기
print(data.text)

## CHECK! 웹 스크래핑이 가능한지 반드시 확인하세요 ##
# 웹 페이지에서 데이터가 보인다고 해서 바로 스크래핑을 실시하지 말고 반드시 소스 코드로 데이터 수집이 가능한지 확인할 것.
# 웹 페이지에서 보이는 페이지 내용과 소스 코드를 실행해 얻은 페이지 내용이 다른 경우, 작성한 소스 코드가 의도한 대로 동작하지 않는 경우가 빈번
# 원하는 정보를 가져오지못하는 이유가 소스 코드 실수 때문이라고 오해하고 문제를 찾는 데 많은 시간을 소모할 수도 있음
```

```python
## 페이징(paging) 처리 : 여러페이지에 걸 담긴 데이터 수집하기
## 요청 url : https://finance.naver.com/item/sise_day.naver?code=005930&page=2
## 쿼리 문자열(동적으로 변경할 수 있는 정보) : URL끝에 물음표(?)로 시작
## 매개변수가 여러 개인 경우, 앰퍼샌드(&)를 사용해 구분

## 페이지 번호 증가용 반복문 작성 ##
# 페이지 번호를 제외한 고정 주소 저장
new_url = 'https://finance.naver.com/item/sise_day.naver?code=005930&page='
```

```python
# 한 페이지를 요청하여 결과 확인
all_tables = pd.DataFrame()
for i in range(1, 2):
    data = requests.get(new_url+str(i), headers=my_headers)
    table = pd.read_html(data.text)[0]
    table.dropna(inplace=True)
    all_tables = pd.concat([all_tables, table])
display(all_tables)
```

```python
## 500일간의 데이터 구하기(50페이지)
all_tables = pd.DataFrame()     # 빈 데이터프레임 선언

for i in range(0,51):  # i가 1~50 까지 반복
    data = requests.get(url+str(i),headers =my_headers)     # new_url에 페이지 번호를 추가하여 웹페이지 요청
    print(f'{i}번째 requests')
    print(new_url + str(i))   # 요청한 url 출력
    table = pd.read_html(data.text)[0]  # 응답내용 테이블을 읽어 저장
    table.dropna(inplace=True)      # 결측치 제거
    all_tables = pd.concat([all_tables,table ])     # all_tables에 읽어온 table 추가
    print(f'{len(table)}줄 추가 / 총 {len(all_tables)}줄')  # 읽어온 데이터 라인수 출력
all_tables
```

```python
# 원본 보존을 위해 다른 이름 할당(전처리용 df생성)
import copy
df = copy.deepcopy(all_tables)
```

## 데이터 시각화

```python
## 한글 열이름을 영문으로 변경하기
# 한글의 경우 그래프를 그리면 깨지는 현상 발생
```

```python
# 열이름 출력하기
df.columns
```

```python
# 열이름 변경하기
# (날짜: date, 종가: end_price, 시가: start_price, 고가: highest_price, 저가: lowest_price)
df.rename(columns={'날짜': 'date', '종가': 'end_price', '시가': 'start_price',
                           '고가': 'highest_price', '저가': 'lowest_price'}, inplace=True)
```

```python
# 데이터 확인하기
df
```

```python
# 불필요한 열 제거하기(전일비, 거래량)
df.drop(['전일비', '거래량'], axis = 1 , inplace=True)
```

```python
# 데이터 확인
df
```

```python
## 최고가와 최저가의 중앙값을 구하여 median_price열 추가하기(최고가와 최저가의 평균)
# ** 교재의 중앙값 구하는 식은 참고할것(최고가와 최저가의 중앙값은 평균이므로 평균으로 구함)
df['median_price'] = (df.highest_price + df.lowest_price)/2
df
```

```python
## ** 참고 : 날짜형식에서 '년월'정보의 문자열로 만들기(강의에서 제외)
all_tables.date.dt.strftime('%Y-%m')  # Y는 네자리연도, y는 두자리 연도
```

```python
###  집계를 위한 달(월) 정보 추가하기 ###
# date열의 월까지만 가져와 month열 만들기
df['month'] = df.date.str[:7]
```
