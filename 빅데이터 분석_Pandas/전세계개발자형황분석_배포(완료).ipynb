#### 3. 많이 사용되는 프로그래밍 언어 그래프 그리기

```python
# 프로그래밍 언어 데이터 추출("LanguageHaveWorkedWith")
lang = languages["LanguageHaveWorkedWith"]
```

```python
# 데이터 확인
lang
```

```python
# 데이터 문자열 변환 후 구분자(;)로 구분
lang.str.split(";")
```

```python
# 리스트로 변환된 항목을 각 행으로 나누기
lang.str.split(';').explode()
```

```python
# 프로그래밍 언어별 응답 수 구하여 lang_list에 할당하기
lang_list = lang.str.split(":").explode().value_counts()
```

```python
# 데이터 확인
lang_list
```

```python
# 상위 개의 언어 파이그래프 그리기
lang_list.nlargest(10).plot.pie()
```

```python
# 백분율 및 제목, y축제목 등 설정하기
lang_list.nlargest(10).plot.pie(ylabel = '',title='language',autopct='%.1f%%')
```

####연령대별 사용 언어 그래프 그리기

```python
# 25~34세 연령대 데이터 추출하여 languages25_34에 할당
languages25_34 = languages[languages.Age == '25-34 years old']
```

```python
# 데이터 확인
languages25_34
```

```python
# 언어의 종류별 개수 구하기
languages25_34 = languages25_34.LanguageHaveWorkedWith.str.split(':').explode().value_counts()
```

```python
# 데이터 확인
languages25_34
```

```python
# 상위 10개 데이터 파이 그래프 그리기
languages25_34.nlargest(10).plot.pie(ylabel='',title='LanguageHaveWorkedWith',autopct="%.1f%%")
```

```python
## 참고 : 그래프 한눈에 보기

import matplotlib.pyplot as plt
plt.figure(figsize=(12,7))
plt.subplot(1,2,1)
plt.pie(lang_list[:10],autopct="%.1f%%", labels=lang_list[:10].index)
plt.title('Languages Used')

plt.subplot(1,2,2)
plt.pie(languages25_34[:10], autopct="%.1f%%", labels=languages25_34[:10].index)
plt.title('Languages for 25~34 years old')

plt.show()
```
