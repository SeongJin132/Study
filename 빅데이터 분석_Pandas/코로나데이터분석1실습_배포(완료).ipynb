# csv파일 탑재하고 정보 알아내기

```python
# csv파일 탑재하기(csv파일을 불러와 data변수에 할당하기)
# --> data라는 데이터프레임 객체가 생성됨
data = pd.read_csv("covid_data.csv")
```

# 필요한 데이터 추출하기

```python
# 필요한 열이름만 추출하기
# (1) 열 내용 추출하기(시리즈)
data["iso_code"]
```

```python
# (2) 여러개의 열 추출하기(데이터프레임)
data[["iso_code","location","total_cases"]]
```

```python
# 하나의 열을 데이터프레임으로 추출하기
data[["iso_code"]]
```

```python
# 분석에 필요한 열만 추출하여 신규 데이터 프레임 생성
covid_df = data[['iso_code', 'location', 'date', 'total_cases','population']]
covid_df
```

```python
# 필요한 행 추출하기(한국과 미국의 데이터만 추출하기)
# (1) '국가명'열에 중복을 없앤 유일한 항목 알아내기
covid_df.location.unique()
```

```python
# (2) 몇개의 나라에 대한 정보가 있을까?('국가명'갯수 확인해보기)
len(covid_df.location.unique())
```

```python
# 국가명 표준코드(iso-code)의 유일한 값 알아내기
covid_df.iso_code.unique()
```

```python
# 한국 데이터만 추출하여 신규 데이터프레임 생성(필요한 행 추출)
kor_df = covid_df[covid_df.iso_code == "KOR"]
```

```python
# 미국 데이터만 추출하여 신규 데이터프레임 생성(필요한 행 추출)
usa_df = covid_df[covid_df.iso_code == "USA"]
```

```python
# 데이터 확인해보기(usa)
kor_df
```

# 결측치 데이터 처리하기

```python
# 결측치 갯수 알아내기
kor_df.isnull().sum()
```

```python
# 대한민국의 결측치 데이터 확인하기(결측치가 있는 행 확인하기)
kor_df[kor_df.isnull()]
```

```python
# 결측치 제거하기
kor_df.dropna(inplace = True)
```

```python
# 미국의 결측치 갯수 알아내기
usa_df.isnull().sum()
```

```python
# 미국의 결측치 제거하기
usa_df.dropna(inplace = True)
```

```python
# 결측치가 제거된 한국의 데이터 확인하기(인덱스?)
kor_df
```

```python
usa_df
```

# 데이터프레임 인덱스(index) 변경하기

```python
# 한국의 인덱스를 날짜로 변경하기
kor_df.set_index("date",inplace = True)
```

```python
# 미국의 인덱스를 날짜로 변경하기
usa_df.set_index("date",inplace = True)
```

# 데이터 시각화

```python
# 한국의 확진자를 그래프로 나타내보기(판다스를 이용)
kor_df["total_cases"].plot(rot = 45)
```

```python
# 데이터 프레임 생성해보기(딕셔너리를 이용한 데이터프레임 생성)
pd.DataFrame({"kor":[1,2,3,4],"usa":[5,6,7,8]})
```

```python
# 한국과 미국의 확진자수로 구성된 데이터프레임 생성하기
# (한국의 인덱스를 기준으로 생성)
kor_usa = pd.DataFrame({'KOR':kor_df.total_cases , 'USA':usa_df.total_cases},index = kor_df.index)
```

```python
# 생성된 데이터프레임 확인하기
kor_usa
```

```python
kor_usa.dropna(inplace = True)
```

```python
# 한국과 미국의 확진자 그래프 생성하기
kor_usa.plot(rot = 45)
```

```python
# 인구대비 확진자수 그래프 생성하기
# (1) 한국의 '확진자수/인구수'를 계산하여 새로운 열 생성하기
#     (선택)반올림하여 소수점 이하 3자리까지 구하기
kor_usa['KOR_RATE'] = round(kor_df.total_cases / kor_df.population,5)
```

```python
# 생성된 데이터 확인해보기
kor_usa
```

```python
# (1) 미국의 '확진자수/인구수'를 계산하여 새로운 열 생성하기
kor_usa['USA_RATE'] = round(usa_df.total_cases / usa_df.population,5)
```

```python
kor_usa
```

```python
# 인구대비 확진자수 그래프 생성하기
kor_usa[["KOR_RATE","USA_RATE"]].plot(rot=45)
```

```python
# 그래프 번위 좁히기(원하는 날짜만 추출하여 표현하기)
kor_usa['2022-01-01':][['KOR_RATE','USA_RATE']].plot(rot=45)
```

```python
kor_usa.to_csv('kor_usa.csv')
```
